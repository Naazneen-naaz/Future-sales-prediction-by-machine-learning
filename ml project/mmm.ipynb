```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Sample data (months, years, and sales)
data = {
    'Months': [4, 8, 12, 16, 20],
    'Years': [2001, 2002, 2003, 2004, 2005],
    'Sales': [100, 200, 300, 400, 500]
}
```

```python
# Create a DataFrame from the data
df = pd.DataFrame(data)
```

```python
plt.scatter(df['Months'], df['Sales'])
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales Data')
plt.show() 
 #Prepare the input features and target variable
X = df[['Months', 'Years']]
y = df['Sales']
```

```python
# Create a linear regression object
model = LinearRegression()
```

```python
# Fit the model to the data
model.fit(X, y)
# Generate predictions for future months
future_months = np.arange(25, 36).reshape(-1, 1)
future_years = np.repeat(2006, 11).reshape(-1, 1)
future_data = np.concatenate((future_months, future_years), axis=1)
predicted_sales = model.predict(future_data)
```

```python
# Create a DataFrame to store the predictions
future_df = pd.DataFrame({'Months': future_months.flatten(), 'Years': future_years.flatten(), 'Predicted Sales': predicted_sales})

print(future_df)
```

```python
# Generate the sales predictions for the whole dataset
predicted_sales_all = model.predict(X)
```

```python
# Plot the data, the linear regression line, and the future predictions
plt.scatter(df['Months'], df['Sales'], label='Actual Sales')
plt.plot(df['Months'], predicted_sales_all, color='red', label='Linear Regression Line')
plt.scatter(future_df['Months'], future_df['Predicted Sales'], color='green', label='Predicted Sales (Future)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales Prediction')
plt.legend()
plt.show()
```
